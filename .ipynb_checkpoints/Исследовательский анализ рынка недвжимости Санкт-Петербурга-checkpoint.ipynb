#  Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. 

**Главная задача проекта** — установить параметры, влияющие на рыночную стоимость объектов недвижимости. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

 **Цель исследования** - установить параметры, влияющие на рыночную стоимость объектов недвижимости.

 **Исходные данные** - данные сервиса «Яндекс.Недвижимость» (архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет).

Вопросы, интересующие заказчика:

- Сколько времени занимает продажа квартиры. Когда продажу можно назвать быстрой / долгой

- Какие факторы больше всего влияют на стоимость квартиры

- Какая средняя стоимость квартиры в 10 населённых пунктов с наибольшим числом объявлений

- Какие факторы влияют на стоимость квартиры в центре Санкт-Петербурга. Насколько они отличаются от данных в целом по городу


Необходимо выявить основные параметры и факторы, влияющие на рыночную цену квартиры. Это позволит отследить аномалии и мошенническую деятельность при автоматизации процесса оценки.

**Ход исследования**

О качестве данных ничего не известно, кроме того, что одна часть информации была указана пользователями самостоятельно, другая - получена автоматически на основе картографических данных (например, расстояние до центра, аэропорта, ближайшего парка и водоёма).

Мы проверим данные на ошибки и оценим их влияние на исследование; затем, на этапе предобработки поищем возможность исправить самые критичные ошибки. Далее проведем расчеты, на основании которых проанализируем данные и выявим параметры, влияющие на ценообразование квартиры.

Таким образом, исследование пройдёт в четыре этапа:

- Обзор данных;

- Исследовательский анализ данных;

# Предобработка данных

Переименуем столбцы  и приведем к одному стилю, для лучшей читаемости и удобства использования в дальнейшем анализе. 

```python
realty_data = realty_data.rename(
    columns={'cityCenters_nearest': 'city_centers_nearest', 
             'parks_around3000':'parks_around_3000', 'ponds_around3000':'ponds_around_3000'}
)
```

```python
realty_data.head(0).T #проверим, что получилось
```

Удалим явные дубликаты, если они есть.

```python
realty_data.duplicated().sum()
```

Явных дубликаов нет.

Посмотрим в каких столбцах есть пропуски и много ли их.

```python
realty_data.isna().sum()
```

Пропусков очень много, выведем эти данные в процентах по каждому столбцу. заполним их где это возможно.

```python
#выведем пропуски в процентном соотношении для наглядности.
pd.DataFrame(realty_data.isna().sum ()/ len(realty_data)*100).style.background_gradient('Pastel1')
```

Пропущенные значения в столбцах 'floors_total'(0,36%) и 'locality_name'(0,2%) можно просто удалить так как их небольшое количество, их удаление не повлияет на результаты исследования.

```python
# удаляем пропуски в столбце 'floors_total' и 'locality_name'.
realty_data.dropna(subset = ['floors_total'], inplace=True)
realty_data.dropna(subset = ['locality_name'], inplace=True)
# изменим тип данных в столбце 'floors_total' на цельночисенный int
realty_data['floors_total'] = realty_data['floors_total'].astype(int)
```

```python
# изменим тип данных в 'last_price' целночисленный int
realty_data['last_price'] = realty_data['last_price'].astype(int)
```

```python
#изменим тип данных в столбце 'first_day_exposition' c object на datetime.
realty_data['first_day_exposition']=pd.to_datetime(realty_data['first_day_exposition'], format="%Y-%m-%dT%H:%M:%S")
realty_data['first_day_exposition'].head() #проверим результат замены типа данных
```

Столбец 'balcony'
Заполним 0 значением, так как если бы он был, по логике, его отметили бы в объявлении.

```python
# заполним пропуски 'balcony' нулевым значением и приведем к int
realty_data['balcony']=realty_data['balcony'].fillna(value=0).astype(int)
#проверим остались ли пропущенные значения
realty_data['balcony'].isna().sum()
```

Пропусков нет.

Столбец 'is_apartment'.
Для выполнения поставленных задач, данные из столбца 'is_apartment' не неужны. Тем не менее тип данных необходимо изменить на bool. Также заполним нулевые значения, значением False исходя из того, что если жильё не является аппартаментами, то человек не будет заполнять эту графу.

```python
# заполним пропуски 'is_apartment' 
realty_data['is_apartment'] = realty_data['is_apartment'].fillna(False)
#проверим остались ли пропущенные значения
realty_data['is_apartment'].isna().sum()
```

Следующий столбец ceiling_height - высота потолков, тип данных float подходит, но столбец имеет нулевые значение, которые необходимо заполнить и пропущенные значения. Пропущенные значения заполним медианными значениями из квартир расположенных в похожих домах, из одного населенного пункта, оставшиеся значения столбца ceiling_height заполним общими медианными значениями.

```python
#Замена пропущенных значений на имеющиеся медианные значения высоты потолков сгруппированные по локации и этажности здания
realty_data['ceiling_height'] = realty_data['ceiling_height'].fillna(
    realty_data.groupby(['locality_name','floors_total'])['ceiling_height'].transform('median'))
#Оставшиеся значения заполним медианными значениями по всей выборке
realty_data['ceiling_height'] = realty_data['ceiling_height'].fillna(realty_data['ceiling_height'].median())
#Проверим количество пропущенных значений
realty_data['ceiling_height'].isna().sum()
```

kitchen_area, living_area - данные, которые могут быть не заполнены просто по той причине, что они необязательны к заполнению, а высчитывать жилую площадь/площадь кухни продавцам некогда. Также выясним, нет ли такой связи, что если квартира - студия, то там отсутствуют эти данные. Ведь в них нет явного деления площади на ту, которая относится к кухне, и ту, которая относится к жилой площади;
Рассмотрим теорию, что именно по студиям не указана эта информация:

```python
#выведем количество пропусков в столбцах kitchen_area и living_area в процентном соотношении
print('Количество NaN по kitchen_area:{:.2f}%'.format(realty_data['kitchen_area'].isnull().sum()/ len(realty_data)*100))
print('Количество NaN по living_area:{:.2f}%'.format (realty_data['living_area'].isnull().sum()/ len(realty_data)*100))
```

```python
# явно высота потолков более 10 м это ошибка в заполнении данных. Поэтому все значения > 19м нужно поделить на 10
def ceiling_height_more_ten(row):
    if row['ceiling_height'] > 19:
        return row['ceiling_height'] * 0.1
    else:
        return row['ceiling_height']
```

Добавление этажа квартиры - первый, последний, другой
Для отнесения этажа квартиры к какой либо группе, напишем функцию, которая будет сравнивать этаж квартиры с количеством этажей в доме и возвращать значение группы.

```python
#добавим столбец со статусом этажа квартиры: первый,последний, другой
#для начала определим функцию, которая будет нам возвращать необходимые значения статуса 
def status_floor(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
rd_filtered['status_floor'] = rd_filtered.apply(status_floor, axis = 1)
```

#  Какие факторы больше всего влияют на общую (полную) стоимость объекта?

 Какие факторы больше всего влияют на общую (полную) стоимость объекта?
  - общей площади;
  - жилой площади;
  - площади кухни;	
  - количество комнат;
  - типа этажа, на котором расположена квартира (первый, последний, другой);
  - даты размещения (день недели, месяц, год);
  Разберёмся какие факторы больше всего влияют на стоимость квартиры. Для этого определим степень корреляции и построим точечные графики. Воспользуемся функцией corr() и посмотрим, как влияют определенные параметры квартиры на её финальную цену.

```python
print('Корреляция  общей площади и цены:',end=':')
print(rd_filtered['total_area'].corr(rd_filtered['last_price']))
rd_filtered.plot(y = 'total_area', x = 'last_price', kind = 'scatter', figsize = (7,5), \
                 grid = True, alpha = 0.1);
```

Можно говорить о зависимости ,если коэффициент корреляции достигает значение (по модулю) от 0.7 до 1. У нас 
"0.82". Зависимость между между площадью квартиры и цены за квартиру - большая.И она прямая. Чем больше площадь, тем больше цена объекта.

```python
print('Корреляция жилой площади и цены',end=':')
print(rd_filtered['living_area'].corr(rd_filtered['last_price']))
rd_filtered.plot(y = 'living_area', x = 'last_price', kind = 'scatter', figsize = (7,5), 
                 grid = True, alpha = 0.1);
```

Здесь тоже наблюдаем сильную прямую зависимость - 0,70. Жилая площадь взаимосвязана с общей площадью, поэтому это неудивительно.

Зависимость между жилой площадью и ценой тоже прослеживается, но меньше. К тому же она положительная, это указывает на прямую зависимость. Думаю это связано с тем. что обычно, чем больше стоимость квартиры, тем больше общая площадь и соответственно жилая площадь.
